```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/netflix_titles.csv')
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.info()
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Number of Movies vs TV Shows')
plt.xlabel('')
plt.ylabel('Count')
plt.show()
print("✅ Most content on Netflix are movies rather than TV shows.")
# Count values
df['type'].value_counts()
```

```python
# Some rows have multiple countries separated by commas — keep the first one
df['main_country'] = df['country'].fillna('Unknown').apply(lambda x: x.split(',')[0])

top_countries = df['main_country'].value_counts().head(10)

plt.figure(figsize=(8,4))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Content-Producing Countries')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
print("✅ USA and India are leading in content production.")

top_countries
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df['release_year'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()

# Count by year
df['release_year'].value_counts().sort_index(ascending=True).tail(10)
```

```python
# Split multiple genres and count the top ones
from collections import Counter

all_genres = []
df['listed_in'].dropna().apply(lambda x: all_genres.extend([genre.strip() for genre in x.split(',')]))

genre_counts = pd.Series(Counter(all_genres)).sort_values(ascending=False).head(10)

plt.figure(figsize=(8,4))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='mako')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()

genre_counts
```

```python
top_directors = df['director'].dropna().value_counts().head(10)

plt.figure(figsize=(8,4))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='cubehelix')
plt.title('Top 10 Most Frequent Directors')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()

top_directors
```

```python
# 5️⃣ Heatmap: Correlation between Numeric Columns
plt.figure(figsize=(6,4))
sns.heatmap(df[['release_year']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
print("✅ Insight: Minimal numeric correlation here, since most columns are categorical.")
```

```python
# 6️⃣ Boxplot Example (Release Year Distribution by Type)
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='type', y='release_year', palette='Set2')
plt.title('Release Year Distribution: Movies vs TV Shows')
plt.xlabel('')
plt.ylabel('Release Year')
plt.show()
print("✅ Insight: TV Shows tend to be slightly newer on average than Movies.")
```
